```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_df = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")

# Display the data table for preview
mouse_df.head()
```

```python
# Checking the number of mice.
mouse_df["Mouse ID"].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = mouse_df.loc[mouse_df.duplicated(["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicatedata = mouse_df.loc[mouse_df["Mouse ID"]=="g989"]
duplicatedata
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mousedf_clean = mouse_df[mouse_df["Mouse ID"].isin(duplicates)==False]
mousedf_clean.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(mousedf_clean["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

tumor_volume = mousedf_clean[["Drug Regimen","Tumor Volume (mm3)"]]
tumor_mean = tumor_volume.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
tumor_median = tumor_volume.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
tumor_var = tumor_volume.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
tumor_std = tumor_volume.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
tumor_sem = tumor_volume.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()

statistics = pd.DataFrame({"Mean Tumor Volume":tumor_mean, "Median Tumor Volume":tumor_median,
                           "Tumor Volume Variance":tumor_var, "Tumor Volume Std. Dev":tumor_std,
                           "Tumor Volume Std. Err":tumor_sem})

statistics
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
aggst = mousedf_clean.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
aggst
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
maxtimepermouse = mousedf_clean.groupby(["Mouse ID"])["Timepoint"].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
tv_df = pd.merge(mousedf_clean, maxtimepermouse, on="Mouse ID")
tv_df = tv_df.rename(columns = {"Timepoint_x":"Timepoint","Timepoint_y":"Max Timepoint"})
tv_df
```

```python
#Dataframe containing only max tumor volumes for each mouse:
volumemax = tv_df.loc[tv_df["Timepoint"]==tv_df["Max Timepoint"],:]
volumemax
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tvdata = []
# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for drug in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumorvolume = volumemax.loc[volumemax["Drug Regimen"]==drug,:]
    
    # add subset
    
    # Determine outliers using upper and lower bounds
    quartiles = tumorvolume["Tumor Volume (mm3)"].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers =[]
    for tumor in tumorvolume["Tumor Volume (mm3)"]:

        if tumor<lower_bound or tumor>upper_bound:
            outliers.append(tumor)
    print(f"{drug} potencial outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
capomulin = volumemax.loc[(volumemax["Drug Regimen"]=="Capomulin"),:]
ramicane = volumemax.loc[(volumemax["Drug Regimen"]=="Ramicane"),:]
infubinol = volumemax.loc[(volumemax["Drug Regimen"]=="Infubinol"),:]
ceftamin = volumemax.loc[(volumemax["Drug Regimen"]=="Ceftamin"),:]

cap_plot = capomulin["Tumor Volume (mm3)"]
ram_plot = ramicane["Tumor Volume (mm3)"]
inf_plot = infubinol["Tumor Volume (mm3)"]
ceft_plot = ceftamin["Tumor Volume (mm3)"]
fig1, ax1 = plt.subplots()
ax1.boxplot([cap_plot, ram_plot, inf_plot, ceft_plot])
ax1.set_xticklabels(treatments)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
x_values = avgvol["Weight (g)"]
y_values = avgvol["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
correlation = st.pearsonr(x_values,y_values)
print(f"The correlation between mouse weight and the average tumor volume is: {round(correlation[0],2)}")
```
